```python
df = pd.concat(
    [headphone ,
    all_appliances ,
    all_electronics ,
    car_accessories ],
    ignore_index=True
)
```

```python
df = df.sample(frac=1 , random_state=42).reset_index(drop=True)
```

```python
df['ratings']=df['ratings'].fillna(0)
```

```python
df['no_of_ratings'] = df['no_of_ratings'].fillna(0)
```

```python
df['discount_price'] = df['discount_price'].fillna(df['actual_price'])
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# df['discount_price'] = df['discount_price'].split('₹')[1]
df['discount_price'] = df['discount_price'].apply(lambda x : x.split('₹')[1])
df['actual_price'] = df['actual_price'].apply(lambda x : x.split('₹')[1])
```

```python
df['discount_price'] = df['discount_price'].str.replace(',' , '').astype('float')
df['actual_price'] = df['actual_price'].str.replace(',' , '').astype('float')
```

```python
df1 = df.copy()
```

```python
df['no_of_ratings'].unique()
```

```python
df['no_of_ratings'] = pd.to_numeric(
    df['no_of_ratings']
        .astype(str)
        .str.replace(',', '', regex=False)
        .str.extract(r'(\d+)')[0],
    errors='coerce'
)

df['ratings'] = pd.to_numeric(
    df['ratings']
        .astype(str)
        .str.replace(',', '', regex=False)
        .str.extract(r'(\d+)')[0],
    errors='coerce'
)
```

```python
df[['no_of_ratings' , 'ratings']] = df[['no_of_ratings' , 'ratings']].fillna(0)
```

```python
df['no_of_ratings'] = df['no_of_ratings'].astype('int')
```

```python
df['ratings'] = df['ratings'].astype('float')
```

```python
df.head(2)
```

```python
df['name'].nunique()
```

```python
df = df.drop_duplicates(subset='name')
```

```python
df['name'].value_counts().head(20)
```

```python
df.shape
```

```python
df['product Id'] = 'P' + (df.index + 1).astype('str')
```

```python
df.head()
```

```python
metadata = df[['product Id' , 'image' , 'link']]
df.drop(columns=['image' , 'link'] , axis=1 , inplace=True)
```

```python
metadata['name'] = df['name']
```

```python
metadata.head()
```

```python
df.info()
```

```python
df['discount percentage'] = (((df['actual_price'] - df['discount_price']) / df['actual_price']) * 100).fillna(0)
```

```python
df.head()
```

```python
df['ratings'].value_counts()
```

```python
df['Rating Bucket'] = np.where(
    df['ratings'] == 0,
    'Unrated',
    pd.cut(
        df['ratings'],
        bins=[0, 2, 3, 4, 5],
        labels=['Low', 'Average', 'Good', 'Excellent']
    )
)
```

```python
df['Popularity Score'] = (
    df['ratings'] * np.log1p(df['no_of_ratings'])
)
```

```python
df['Value Score'] = (
    df['Popularity Score'] / df['discount_price']
).fillna(0.0)
```

```python
noise_words = [
    "premium",
    "original",
    "latest",
    "best",
    "new",
    "quality"
]

def clean_text(name):
    name = name.lower()
    for word in noise_words:
        name = name.replace(word , "")

    return name
df['cleaned_name'] = df['name'].apply(lambda name : clean_text(name))
```

```python
df['main_category'] = df['main_category'].str.replace("tv, audio & cameras" , "tv audio camera")
```

```python
df['cleaned_sub_category'] = (
    df['sub_category']
    .str.replace("All Electronics", "Electronics", regex=False)
    .str.replace("All Appliances", "Appliances", regex=False)
)
```

```python
df.head()
```

```python
df['cleaned_name'] = (
    df['cleaned_name']
    .str.replace('...','')
    .str.replace('..','')
    .str.replace('.','')
    .str.replace('-','')
    .str.replace('|','')
    .str.replace('||','')
    .str.replace('/','')
    .str.replace('+','')   
)
```

```python
df['tags'] = (df['main_category'] + " "  + df['sub_category'] + " " + df['name'])
```
